```python
# linspace: Conjunto de números equiespaciados, analizar los ejes o gráficos (NO TIENE QUE VER CON LOS EJES DE LA GRÁFICA)

import numpy as np

valores = np.linspace (0, 100, 50)
valores
```

```python
nuevos_valores = np.where (valores < 30)           # Indices de mi arreglo donde se cumple la condición
nuevos_valores
```

#Gráfica

*ESTE ES UN ERROR PORQUE SOLO ALMACENA EN UNA ÚNICA VARIABLE*

```python
import numpy as np

X = np.linspace (-10,10, 1000)    # Smooth para darle mas curva a la gráfica,
                                  # X es un arreglo de valores

Y = math.log (X + math.sin (math.pi/2)) + math.sqrt (math.tan ((X/5) + math.pi/8))
```

*ALMACENA LA RESPEUSTA DE CIERTA CANTIDAD DE VALORES EN UNA SERIE DE NÚMEROS*

```python
import numpy as np

X = np.linspace(0, 10, 100)

Y = np.log(X + np.sin(math.pi/2)) + np.sqrt(np.arctan((X/5) + np.pi/8))
Y
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt

X = np.linspace(-10, 10, 1000)
Y = np.log(np.abs(X + np.sin(math.pi/2))) + np.sqrt(np.abs(np.arctan((X/5) + np.pi/8)))

myGraph = plt.figure(figsize=(10, 5))
plt.plot(X, Y, color='purple', label='Y')  # Agregar etiqueta para la línea
plt.title("Gráfica X, Y")
plt.grid(True)
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.legend()  # Mostrar la leyenda
plt.show()
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

X = np.linspace(-10, 10, 100)
Y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(X, Y)
Z = np.log(np.abs(X + np.sin(math.pi/2))) + np.sqrt(np.abs(np.arctan((Y/5) + np.pi/8)))

myGraph = plt.figure(figsize=(10, 7))
ax = myGraph.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis')

plt.title("Gráfica 3D de X, Y, Z")
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')

plt.show()
```
